```python
import random
import pandas as pd
df=pd.read_csv("/content/answer.csv",sep="\t")

def add_random_words(text):
    random_words = ["উদাহরণ", "পরীক্ষা", "অতিরিক্ত", "শব্দ", "যোগ"]
    if pd.isna(text) or len(text.split()) <= 1:
        additional_words = " ".join(random.choices(random_words, k=2))
        return f"{text} {additional_words}".strip()
    return text
df["Translation"] = df["Translation"].apply(add_random_words)
df.to_csv("/content/ans.csv", sep="\t", index=False)
print("Saved file to /content/ans.csv")
```

```python
#To convert comma separated to tab separated
import csv
input_file = "/content/answer_hi.csv"
output_file = "answer_tab_hin.csv"

with open(input_file, "r", encoding="utf-8") as csvfile, \
    open(output_file, "w", encoding="utf-8", newline="") as tsvfile:

    reader = csv.reader(csvfile, delimiter=",")
    writer = csv.writer(tsvfile, delimiter="\t")

    for row in reader:
        writer.writerow(row)

print("conversion done: ", output_file)
```
